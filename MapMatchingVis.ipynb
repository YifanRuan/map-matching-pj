```python
import folium
import pandas
```

```python
f = open("improve2_raw_traj.txt")
kM = int(f.readline());
traj_points = []
for i in range(kM):
    x = []
    while True:
        tmp = f.readline().split();
        if len(tmp) == 1:
            break;
        tmp[0] = int(tmp[0])
        tmp[1] = float(tmp[1])
        tmp[2] = float(tmp[2])
        x.append(tmp)
    traj_points.append(x)
    
def find_route(num):
    location = []
    for loc in traj_points[num]:
        location.append([loc[1], loc[2]])
    return location
```

```python
sample_in = open("sample.in")
edge_num = int(sample_in.readline())
edge = []
for i in range(edge_num):
    edge.append(sample_in.readline().split())
```

```python
my_f = open("my.out")
ans_f = open("improve2_label.txt")
my = []
ans = []
my_f.readline()
ans_f.readline()
for i in range(kM):
    my.append(list(map(int, my_f.readline().split())))
    ans.append(list(map(int, ans_f.readline().split())))
```

```python
# 307, 20, 44, 162, 202
num = 202

locs = find_route(num)
m = folium.Map(locs[len(locs) // 2], zoom_start = 15)
route = folium.PolyLine(locs).add_to(m)
mk = folium.Marker(locs[0]).add_to(m)
route_edge = []

def draw_edge(label, c = 'red'):
    list1 = edge[label][6::2]
    list1 = list(map(float, list1))
    list2 = edge[label][7::2]
    list2 = list(map(float, list2))
    route_edge = folium.PolyLine(zip(list1, list2), color = c).add_to(m)
    
def draw_route(numx):
    for i in ans[numx]:
        draw_edge(i, 'green')
    for i in my[numx]:
        draw_edge(i)
        
draw_route(num)
m
```
